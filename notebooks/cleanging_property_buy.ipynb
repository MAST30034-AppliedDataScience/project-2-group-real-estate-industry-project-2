```python
# Remove all commas after number
ddf['rented_prices_cleaned'] = ddf['rented_prices_cleaned'].str.replace(r'(?<=\d),', '', regex=True)
```

```python
# add " ' " in front of the "$" sign if there is no " ' " in front of the "$" sign
ddf['rented_prices_cleaned'] = ddf['rented_prices_cleaned'].apply(lambda x: re.sub(r'(?<!\')\$', '\'', x))

# add " ' " in front of the "}" sign if the above condition is met
ddf['rented_prices_cleaned'] = ddf['rented_prices_cleaned'].apply(lambda x: re.sub(r'(?<!\')\}', '\'', x))

# add " } " behind the "0'" if "{" is present but the next curly braces is not "}"
ddf['rented_prices_cleaned'] = ddf['rented_prices_cleaned'].apply(lambda x: re.sub(r"(?<=0')(?=[^}]*$)", "}", x))
```

```python
ddf_list = ddf[['rented_prices_cleaned']].values.tolist()
ddf_list
```

```python
# add all the matches to a list and convert it to a dataframe
matches_list = []
for i in range(len(ddf_list)):
    for j in range(len(ddf_list[i])):
        matches = re.findall(r"'date': '([^']*)', 'price': '([^']*)'", ddf_list[i][j])
        matches_list.append(matches)

len(matches_list)
```

```python
# Flattening the nested tuples to create a single list of tuples (date, price)
flat_data = [(item[0], item[1]) for sublist in matches_list for item in sublist]

# Creating DataFrame with two columns: 'date' and 'price'
df = pd.DataFrame(flat_data, columns=['date', 'price'])

# Displaying the DataFrame
df
```
